# 1. 인구 피라미드

## 남자 데이터 정의

```python
import pandas as pd
file_name ='202201_202201_연령별인구현황_월간.xlsx'
df_m =pd.read_excel(file_name,skiprows=3,index_col='행정기관',usecols='B,E:Y')
df_m.head(3)
```

```python
df_m.iloc[0]=df_m.iloc[0].str.replace(',','').astype(int)  # 1,195,951 -> 1195951 (정수형)
```

```python
df_m.iloc[0]
```

## 여자 데이터 정의

```python
df_w =pd.read_excel(file_name,skiprows=3,index_col='행정기관',usecols='B,AB:AV')
df_w.head(3)
```

```python
df_m.columns
```

```python
df_w.columns
```

```python
df_w.columns=df_m.columns  # 컬럼명 동일하게 만들기 
```

```python
df_w.columns
```

```python
df_w.iloc[0]=df_w.iloc[0].str.replace(',','').astype(int)  # 1,125,896 -> 1125896 (정수형)
```

```python
df_w.iloc[0]
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['font.family']="Malgun Gothic" # windows
matplotlib.rcParams['font.size']=15 # 글자크기 
matplotlib.rcParams['axes.unicode_minus']=False # 한글 폰트 사용시 ,마이너스 글자 깨짐현상 해결
```

```python
plt.barh(df_m.columns,-df_m.iloc[0]//1000 )  # 단위 천 명 
plt.barh(df_w.columns,df_w.iloc[0]//1000 ) # 단위 천 명
plt.title('2022년 대한민국 인구 피라미드')
plt.savefig('2022_인구피라미드.png')
plt.show()
```
