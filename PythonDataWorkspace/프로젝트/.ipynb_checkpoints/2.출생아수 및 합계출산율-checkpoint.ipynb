# 2. 출생아 수 및 합계출산율

```python
import pandas as pd 
file_name='stat_142801.xls'
df=pd.read_excel(file_name,skiprows=2,nrows=2,index_col=0)
df
```

```python
df.index.values
```

```python
df.rename(index={'출생아\xa0수':'출생아 수','합계\xa0출산율':'합계 출산율'},inplace=True)
```

```python
df.index.values
```

```python
df.loc['출생아 수']  # location 문자열 사용 
```

```python
df.iloc[0]   # integer location  숫자 사용 
```

```python
df
```

```python
df=df.T  # row랑 column 변경 
```

```python
df
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['font.family']="Malgun Gothic" # windows
matplotlib.rcParams['font.size']=15 # 글자크기 
matplotlib.rcParams['axes.unicode_minus']=False # 한글 폰트 사용시 ,마이너스 글자 깨짐현상 해결
```

```python
plt.plot(df.index,df['출생아 수'])
plt.plot(df.index,df['합계 출산율'])
plt.show()
```

```python
fig,ax1 = plt.subplots(figsize=(7,5))
ax1.plot(df.index,df['출생아 수'],color='#ff812d')

ax2 = ax1.twinx()  # x축을 공유하는 쌍둥이 axis 
ax2.plot(df.index,df['합계 출산율'],color='#ffd100')
plt.show()
```

```python
fig,ax1 = plt.subplots(figsize=(12,5))
ax1.set_ylabel('출생아 수(천 명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'],color='#ff812d')
for idx,value in enumerate(df['출생아 수']):
    ax1.text(idx,value +12,value,ha='center')

ax2 = ax1.twinx()  # x축을 공유하는 쌍둥이 axis 
ax2.set_ylabel('합계 출산율(가임 여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'],color='#ffd100')
plt.show()
```

```python
fig,ax1 = plt.subplots(figsize=(12,5))
ax1.set_ylabel('출생아 수(천 명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'],color='#ff812d')
for idx,value in enumerate(df['출생아 수']):
    ax1.text(idx,value +12,value,ha='center')

ax2 = ax1.twinx()  # x축을 공유하는 쌍둥이 axis 
ax2.set_ylabel('합계 출산율(가임 여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'],color='#ffd100',marker='o',ms=12,lw=5,markeredgecolor='w',markeredgewidth=3)
plt.show()
```

```python
fig,ax1 = plt.subplots(figsize=(12,5))
ax1.set_ylabel('출생아 수(천 명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'],color='#ff812d')
for idx,value in enumerate(df['출생아 수']):
    ax1.text(idx,value +12,value,ha='center')

ax2 = ax1.twinx()  # x축을 공유하는 쌍둥이 axis 
ax2.set_ylabel('합계 출산율(가임 여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'],color='#ffd100',marker='o',ms=12,lw=5,markeredgecolor='w',markeredgewidth=3)
for idx,value in enumerate(df['합계 출산율']):
    ax2.text(idx,value +0.08,value,ha='center')
plt.show()
```

```python
fig,ax1 = plt.subplots(figsize=(12,5))
fig.suptitle('출생아 수 및 합계 출산율')
ax1.set_ylabel('출생아 수(천 명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'],color='#ff812d')
for idx,value in enumerate(df['출생아 수']):
    ax1.text(idx,value +12,value,ha='center')

ax2 = ax1.twinx()  # x축을 공유하는 쌍둥이 axis 
ax2.set_ylabel('합계 출산율(가임 여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'],color='#ffd100',marker='o',ms=12,lw=5,markeredgecolor='w',markeredgewidth=3)
for idx,value in enumerate(df['합계 출산율']):
    ax2.text(idx,value +0.08,value,ha='center')
plt.savefig('출생아 수 및 합계출산율.png')
plt.show()
```
