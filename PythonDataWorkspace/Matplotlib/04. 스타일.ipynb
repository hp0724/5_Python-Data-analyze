# 4.스타일

```python
import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['font.family']="Malgun Gothic" # windows
matplotlib.rcParams['font.size']=15 # 글자크기 
matplotlib.rcParams['axes.unicode_minus']=False # 한글 폰트 사용시 ,마이너스 글자 깨짐현상 해결
```

```python
x= [1,2,3]
y= [2,4,8]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,linewidth=5)
plt.show()
```

## 마커 (marker)
matplotlib.marker 구글 참조

```python
plt.plot(x,y,marker='o')
plt.show()
```

```python
plt.plot(x,y,marker='o',linestyle='None')
plt.show()
```

```python
plt.plot(x,y,marker='v')
plt.show()
```

```python
plt.plot(x,y,marker='v',markersize=10)
plt.show()
```

```python
plt.plot(x,y,marker='X',markersize=10)
plt.show()
#matplotlib.markers 참조 
```

```python
plt.plot(x,y,marker='o',markersize=5,markeredgecolor='red')
plt.show()
```

```python
plt.plot(x,y,marker='o',markersize=5,markeredgecolor='red',markerfacecolor='yellow')
plt.show()
```

```python
plt.plot(x,y,linestyle='-.')
plt.show()
```

```python
plt.plot(x,y,linestyle='-')
plt.show()
```

## 포맷

```python
plt.plot(x,y,'ro--') # color ,marker, linke style 
plt.show()
```

```python
plt.plot(x,y,'go')
plt.show()
```

## 축약어 
- 참조 https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html 

```python
plt.plot(x,y,marker='o',mfc='red',ms=10,mec='blue',ls=':',c='blue')
plt.show()
```

## 투명도

```python
plt.plot(x,y,marker='o',mfc='red',ms=10,alpha=0.3) # alpha 투명도 0~1
plt.show()
```

## 그래프 크기

```python
plt.figure(figsize=(4,2))
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(10,5),dpi=50)  #dot per inch 확대 
plt.plot(x,y)
plt.show()
```
