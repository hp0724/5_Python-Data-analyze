# 13.원 그래프(기본)

```python
import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['font.family']="Malgun Gothic" # windows
matplotlib.rcParams['font.size']=15 # 글자크기 
matplotlib.rcParams['axes.unicode_minus']=False # 한글 폰트 사용시 ,마이너스 글자 깨짐현상 해결
```

```python
values=[30,25,20,13,10,2]
labels = ['Python','Java','Javascript','C#','C/C++','ETC']
# values랑 labels 크기 같아야함 
plt.pie(values,labels=labels,autopct='%.1f%%',startangle=180,counterclock=False) 
plt.show()
```

```python
values=[30,25,20,13,10,2]
labels = ['Python','Java','Javascript','C#','C/C++','ETC']
#explode = [0.3,0.3,0.1,0.1,0.1,0.1]
explode =[0.05]*len(labels)
plt.pie(values,labels=labels,explode=explode)
plt.show()
```

```python
plt.pie(values,labels=labels,explode=explode)
plt.legend(loc=(1.2,0.3))
plt.show()
```

```python
plt.pie(values,labels=labels,explode=explode)
plt.legend(loc=(1.2,0.3),title='언어')
plt.title('언어별 선호도')
plt.show()
```
